```python
lineup = sb.lineups(match_id=match_id)['England']
played_lineup = lineup[lineup['positions'].apply(lambda x: len(x) != 0)]
played_lineup
```

```python
all_passes['pass_outcome'].isnull()
```

```python
def drawPassPlot(match_id, player_id, player_name):
    passes = sb.events(match_id=match_id).query(f'player_id=={player_id} & type=="Pass"')

    df1 = pd.DataFrame(passes["location"].tolist(), columns=['location_x', 'location_y'])
    # locationの補正
    df1['location_x'] = df1['location_x'] * 105 / 120
    df1['location_y'] = 68 - df1['location_y'] * 68 / 80
    
    df2 = pd.DataFrame(passes["pass_end_location"].tolist(), columns=['pass_end_location_x', 'pass_end_location_y'])
    # locationの補正
    df2['pass_end_location_x'] = df2['pass_end_location_x'] * 105 / 120
    df2['pass_end_location_y'] = 68 - df2['pass_end_location_y'] * 68 / 80

    print(player_name)
    if len(df1) > 0:
        ax = matplotsoccer.actions(
            location=pd.concat([df1, df2], axis=1),
            action_type=passes.type.map(lambda x: 'pass'), # 'Pass'だと表示が四角になってしまう...
            result=all_passes['pass_outcome'].isnull(),
            zoom=False,
            figsize=6,
            show_legend=False
        )
    else:
        print('no passes')
        print('')
```

```python
for index, player in played_lineup.iterrows():
    drawPassPlot(match_id, player.player_id, player.player_name)
```

```python
grouped_england_events = sb.events(match_id=match_id).query('team=="England"').groupby('type').size() #.sort_values(ascending=False)
grouped_spain_events = sb.events(match_id=match_id).query('team=="Spain"').groupby('type').size()# .sort_values(ascending=False)
grouped_events = pd.concat([grouped_england_events, grouped_spain_events], axis=1).fillna(0)
grouped_events.columns = ['England', 'Spain']
grouped_events['England'] = grouped_events['England'].astype(int)
grouped_events
```

```python
spain_lineup = sb.lineups(match_id=match_id)['Spain']
spain_played_lineup = spain_lineup[spain_lineup['positions'].apply(lambda x: len(x) != 0)]

for index, player in spain_played_lineup.iterrows():
    drawPassPlot(match_id, player.player_id, player.player_name)
```
